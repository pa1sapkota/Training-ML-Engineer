# Builder Design Pattern 
separate the construction of a complex object from its representation so that you can use the same construction process to create different representations.

- similar to the factory design pattern as they create the object on the runtime 

Terminology
- Product: The Product being built.
- Builder: Builds the concrete product. Implements the Builder Interface.
- Builder Interface: The Interface that the Concrete builder should implement.
- Director: Has a construct() method that when called creates a customized product.
